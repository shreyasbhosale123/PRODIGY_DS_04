```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("/content/twitter_training.csv", header=None)
df.head()
```

```python
df.columns = ["ID","Topic","Sentiment","Tweet"]
df.head()
```

```python
sns.countplot(x="Sentiment", data=df)

plt.title("Sentiment Distribution in Social Media Data")
plt.show()
```

```python
plt.figure(figsize=(10,6))

sns.countplot(x="Topic", hue="Sentiment", data=df)

plt.title("Sentiment by Topic")
plt.xticks(rotation=90)

plt.show()
```
